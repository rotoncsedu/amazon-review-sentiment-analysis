#### <span style='color:#27AE60; font-weight:bold;font-size:24px; text-shadow:0 0 3px rgba(15,52,96,0.6);'>Step 4 — Prepare Labels</span>

```python
df = df[df['Positive'] != -1].copy()
df['label'] = df['Positive'].map({1: 'Positive', 0: 'Negative'})

print("Label distribution:")
print(df['label'].value_counts())
print()
print("Sample Positive review:")
print(df[df['label'] == 'Positive']['reviewText'].iloc[0])
print()
print("Sample Negative review:")
print(df[df['label'] == 'Negative']['reviewText'].iloc[0])
```
